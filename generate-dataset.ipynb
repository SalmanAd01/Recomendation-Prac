```python
import requests as req
from pprint import pprint as pp
url = "http://arena.siesgst.ac.in/api/problems"
headers ={
    "Accept": "application/vnd.arena+json;version=1",
}
res = req.get(url, headers =headers)
pp(res.json())
```

```python
lenght = len(res.json())
print(lenght)
```

```python
import pandas as pd
df = pd.DataFrame(res.json())
df.head()
```

```python
df.isnull().sum()
```

```python
df["tags"] 
```

```python
for i in range(len(df["tags"])):
    if len(df["tags"][i]) == 0:
        df["tags"][i].append("math")
```

```python
for i in range(len(df["tags"])):
    # list to string
    df["tags"][i] = str(df["tags"][i])
    # remove brackets
    df["tags"][i] = df["tags"][i].replace("[","")
    df["tags"][i] = df["tags"][i].replace("]","")
    df["tags"][i] = df["tags"][i].replace(","," ")
    df["tags"][i] = df["tags"][i].replace("'","")
    df["tags"][i] = df["tags"][i].replace("#","")

    
```
